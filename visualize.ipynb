# VISUALIZE FEATURES AND FACE

This code is used mostly for analyzing photo and detecting faces and extracting features

```python
from functions import *
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("src/Thiago.jpeg")
face = detect_face(image)
features = extract_features(image)

image_with_face = image.copy()
x1, y1, x2, y2 = face
cv2.rectangle(image_with_face, (x1, y1), (x2, y2), (0, 255, 0), 2)

plt.imshow(cv2.cvtColor(image_with_face, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()

print(f'Features: {features}')
```
